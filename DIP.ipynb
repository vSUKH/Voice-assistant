```python
# Digit Multiplication Puzzle-----------------------------------
num = int(input("whats the num"))
n = len(str(num))
s = 1
product = 1
while s <= n:
    digit = num%10 * (1)
    product *= digit
    num //= 10
    s+=1
print(product)
```

```python
# Paliandrome--------------------------------------------------
value = input("Enter a string or number: ")

start = 0
end = len(value) - 1

while start < end:
    if value[start] != value[end]:  # Check if characters are the same
        print(f"'{value}' is not a palindrome.")
        break
    start += 1
    end -= 1
else:
    print(f"'{value}' is a palindrome.")
```
